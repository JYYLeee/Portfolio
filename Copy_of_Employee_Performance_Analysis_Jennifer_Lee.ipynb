The "Employee_Performance" dataset comprises the following variables:

- **Work_Hours_per_Week:** Hours worked by the employee per week.
- **Salary:** Salary of the employee.
- **Performance_Rating:** Employee performance rating.
- **Department:** Department in which the employee works (Sales, Marketing, Engineering, Finance, Human Resources).

## **Question 1:** [How is the Distribution of the total employee performance ratings?]

**Reason for Choosing [Visualization Type]:**  
I chose a [Bar Graph] for this question because... It is easier to track distributions with the visual. There is also an option for a trend line.  


```python
# Run Your Statistics

print(employee_df[['Employee_ID', 'Performance_Rating']])
```

```python
# Run Your Visualization
plt.figure(figsize=(10, 6))
sns.histplot(data=employee_df, x='Performance_Rating', kde=True)


plt.title('Distribution of Performance Ratings')
plt.xlabel('Employee Performance Ratings')
plt.ylabel('Frequency')
```

## **Question 2:** [Is there a relationship between Salary and Performance Rating?]

**Reason for Choosing [Scatter Plot]:**  
I chose a Scatter plot because it is the best model for finding correlations between two variables. There is also an overall trend line to help show an overview.


```python
# Run Your Statistics
print(employee_df[['Employee_ID','Salary', 'Performance_Rating']])
```

```python
# Run Your Visualization
plt.figure(figsize=(10, 6))

sns.regplot(data=employee_df,x='Salary',y='Performance_Rating', line_kws={'color':'red'},)


plt.title('Relationship between Salary and Performance rating')
plt.xlabel('Salary')
plt.ylabel('Performance Rating')
# Clean up final result
sns.despine()
#line_kws changes color of best fit line.
#upward trend
#Shows that typically the more Salary you have, the higher the performance rating.
```

Based on the visualization, we can observe that...
